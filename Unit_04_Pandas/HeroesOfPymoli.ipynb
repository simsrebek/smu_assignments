## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Purchasing Analysis Total
#Initializing data for dataframe 

unique_items_totl = Purchase_Data["Item ID"].nunique()
avg_price_totl = Purchase_Data["Price"].mean()
revenue_totl = Purchase_Data["Price"].sum()
nbr_purch_totl = Purchase_Data["Purchase ID"].nunique()



#Creating data frame to hold results
#Passing dictionary to data frame and formatted

sum_dict = [{"Number of Unique Items":unique_items_totl,
                          "Average Price":avg_price_totl,
                          "Number of Purchases":nbr_purch_totl,
                          "Total Revenue":revenue_totl}]


sum_totl = pd.DataFrame(sum_dict)
sum_totl["Total Revenue"] = sum_totl["Total Revenue"].map("${:.2f}".format)
sum_totl["Average Price"] = sum_totl["Average Price"].map("${:.2f}".format)
Purchase_Analysis_totl = sum_totl[["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]]

#Display DataFrame
Purchase_Analysis_totl

```


## Purchasing Analysis (Gender)

```python
#Purchasing Analysis by Gender

#Created variables for gender with groupby
purchase_count_gen = Purchase_Data["Gender"].value_counts()
average_price_gen = round(Purchase_Data.groupby("Gender")["Price"].mean(),2)
purchase_value_gen = round(Purchase_Data.groupby("Gender")["Price"].sum(),2)
uniq_purchase_cnt = Purchase_Data.groupby("Gender")["SN"].nunique()

#Build DataFrame and format
gender_purchase_analysis = pd.DataFrame({"Purchase Count": purchase_count_gen,
                                        "Average Purchase Price": average_price_gen.map("${:.2f}".format),
                                        "Total Purchase Value": purchase_value_gen.map("${:.2f}".format),
                                        "Average Purchase Total per Person": round(purchase_value_gen/uniq_purchase_cnt,2).map("${:.2f}".format)
                                        })
#Display summary df
gender_purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Age Demographics

#Create Bins for age demographics
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]


#Establish variables for count and percent
Purchase_Data["Age Demographics"] = pd.cut(Purchase_Data["Age"], age_bins, labels=bin_names)
cnt_age_dem = Purchase_Data.groupby("Age Demographics")["SN"].nunique()
pct_age_dem = Purchase_Data.groupby("Age Demographics")["SN"].nunique().sum()


#Build Dataframe and format                                             
age_demographics = pd.DataFrame({"Total Count": cnt_age_dem,
                                 "Percentage of Players": round(cnt_age_dem/pct_age_dem*100,2).map("{:.2f}%".format)})
    
#display summary df
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchasing Analysis by Age

#Create variables and categories
purch_count_age = Purchase_Data.groupby("Age Demographics")["Purchase ID"].count()
avg_price_age = Purchase_Data.groupby("Age Demographics")["Price"].mean()
totl_pur_value_age = Purchase_Data.groupby("Age Demographics")["Price"].sum()
totl_purch_age = Purchase_Data.groupby("Age Demographics")["SN"].nunique()

#Build Dataframe and format
purchase_analysis_age = pd.DataFrame({"Purchase Count": purch_count_age,
                                      "Average Purchase Price": round(avg_price_age,2).map("${:.2f}".format),
                                      "Total Purchase Value": totl_pur_value_age.map("${:.2f}".format),
                                      "Avg Total Purchase per Person": round(totl_pur_value_age/totl_purch_age,2).map("${:.2f}".format)})
                         
#Display summary df
purchase_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Top spenders per user - This was more difficult than I thought. The 'most popular items' was less complicated. 
#I am open to suggestions as to how to be more effecient in this section. yeesh.

#Create variables using grouby for 'SN' userID
top_spend = Purchase_Data.groupby(["SN"])
top_spend_count = top_spend["Item ID"].count()
top_spend_total = top_spend["Price"].sum()
top_spend_avg = (top_spend_total / top_spend_count)

# Build summary DataFrame
top_spenders = pd.DataFrame({"Purchase Count": top_spend_count,
                             "Average Purchase Price": top_spend_avg,  
                             "Total Purchase Value": top_spend_total})

#Sort and Format
top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False) 
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:.2f}".format)
top_spenders = top_spenders[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

#Display df
top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Most popular items

#Create variables - DataFrame and format
popular_item  = Purchase_Data.groupby(["Item ID", "Item Name"])["Price"].count()
popular_price = Purchase_Data.groupby(["Item ID", "Item Name"])["Price"].mean()
popular_total = Purchase_Data.groupby(["Item ID", "Item Name"])["Price"].sum()

most_popular_items = pd.DataFrame({"Purchase Count":popular_item,
                                   "Item Price": popular_price.map("${:.2f}".format),
                                   "Total Purchase Value": popular_total.map("${:.2f}".format)})

#Sort and view summary data frame
most_popular_items.sort_values(by="Purchase Count",ascending= False).head()
```
